```python
import tensorflow as tf
import numpy as np
```

```python
xs = np.arange(1,10).astype('float')
ys = xs * 2 + 1
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 1, input_shape = [1])
])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
model.fit(xs, ys, epochs = 100)
```

```python
model.predict([10])
```
